Answer 6.1.1,6.1.2 > Read csv file with pandas and pick the column

```python
import pandas as pd

df = pd.read_csv("data_banknote_authentication.csv")

X = df.drop(df.columns[-1], axis=1)
y = df['class']
```

Answer 6.1.3 > Training and testing data selective

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.20, random_state=20)
```

Answer 6.1.4,6.1.5 > Import linear model and predict

```python
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, classification_report

svm = SVC(kernel='linear')
svm.fit(X_train, y_train)

y_pred_linear = svm.predict(X_test)

print(f"A:\n{confusion_matrix(y_test, y_pred_linear)}")
print(f"B:\n{classification_report(y_test, y_pred_linear)}")
```

Answer 6.1.6 > Import rbf model and replete the predict again

```python
rbf_svm = SVC(kernel='rbf')
rbf_svm.fit(X_train, y_train)

y_pred_rbf = rbf_svm.predict(X_test)

print(f"A:\n{confusion_matrix(y_test, y_pred_rbf)}")
print(f"B:\n{classification_report(y_test, y_pred_rbf)}")
```

Answer 6.1.7 > My conclusion
The RBF Kernel achieves higher accuracy compared to the linear Kernel.But the linear and RBF kernel a ble to classification in linear.

Answer 6.2.1,6.2.2 > Read the csv file and select X(Feature)and y(Target)

```python
df2 = pd.read_csv('weight-height.csv')

X = df2[['Height']]*2.54
y = df2['Weight']*0.45359237
```

Answer 6.2.3 > Testing and training 80/20

```python
X1_train, X1_test, y1_train, y1_test = train_test_split(X,y,test_size=0.20, random_state=42)
```

Answer 6.2.4 > Prepare scaler Normalization and Standardization

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

min_max_scaler = MinMaxScaler()
X1_train_norm = min_max_scaler.fit_transform(X1_train)
X1_test_norm = min_max_scaler.transform(X1_test)

std_scaler = StandardScaler()
X1_train_std = std_scaler.fit_transform(X1_train)
X1_test_std = std_scaler.transform(X1_test)
```

Answer 6.2.5 > Training and Predict in Unscaled

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score

knn_un = KNeighborsRegressor(n_neighbors=5)
knn_un.fit(X1_train, y1_train)
y1_pred = knn_un.predict(X1_test)
r2_unscaled = r2_score(y1_test, y1_pred)
```

Answer 6.2.6 > Training and Predict in Normalization

```python
knn_norm = KNeighborsRegressor(n_neighbors=5)
knn_norm.fit(X1_train_norm, y1_train)
y1_pred_norm = knn_norm.predict(X1_test_norm)
r2_norm = r2_score(y1_test, y1_pred_norm)
```

Answer 6.2.7 > Training and Predict in Standardization

```python
knn_std = KNeighborsRegressor(n_neighbors=5)
knn_std.fit(X1_train_std, y1_train)
y1_pred_std = knn_std.predict(X1_test_std)
r2_std = r2_score(y1_test, y1_pred_std)
```

```python
print(f"R square Score (Unscaled Data):     {r2_unscaled:.25f}")
print(f"R square Score (Normalized Data):   {r2_norm:.25f}")
print(f"R square Score (Standardized Data): {r2_std:.25f}")
```
